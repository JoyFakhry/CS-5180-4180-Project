```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def plot_y(y, title="", xlabel="", ylabel=""):
    ax = plt.axes()
    X = [i for i in range(len(y))]
    ax.plot(X, y)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
```

```python
rewards = np.genfromtxt("rewards.csv", delimiter = "\t")
```

```python
fig = plt.figure("rewards vs timestep")
plot_y(rewards, "rewards for random policy(timestep)", "steps", "rewards")
fig = plt.figure("rewards vs episode")
# plot_y(rewards, "rewards for random policy(episode)", "steps", "rewards")
fig = plt.figure("length vs episode")
# plot_y(rewards, "length of episode for random policy", "steps", "rewards")
```
